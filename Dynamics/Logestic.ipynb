```python
%matplotlib inline

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import matplotlib.pyplot as plt


import numpy as np
```

```python
def logestic_plot(R: float, x0: float) -> None:
    n = 100
    x = np.linspace(0, n, n)
    y = np.zeros([n])
    y[0] = x0
    for i in range(0, n):
        if not i == 0:
            y[i] = R*(y[i-1])*(1-y[i-1])
    
    # print(f'type of x {x} type of y {y}')
    # print(x)
    avg = np.mean(y[60:100])
    print(avg)
    avg_list = [avg for i in range(n)]
    plt.scatter(x, y)
    plt.plot(x, y)
    plt.plot(x, avg_list, c='r')
    plt.xlim(0, n)
    plt.ylim(-1, 1)
    plt.grid(True)
    plt.show()
interact(logestic_plot, R=(1, 5, 0.1), x0=(0, 1, 0.01))
```

```python
logestic_plot(3.4, 0.6)
```
